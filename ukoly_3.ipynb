```python
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        c = a + b
        a = b
        b = c

for cislo in fibonacci(10):
    print(cislo)
```

```python
suda_cisla_gen = (x for x in range(100) if x % 2 == 0)

for c in suda_cisla_gen:
    print(c, end=" ")
```

```python
def test(n):
    if n < 2:
        return False
    gen = (n % d != 0 for d in range(2, n))
    return all(gen)
print("\nJe 7 prvočíslo?", test(7))
print("Je 10 prvočíslo?", test(10))
```
